```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("stack-overflow-developer-survey-2025/survey_results_public_without_SO.csv")
```

<h2>Line Charts zur Visualiserung der Daten</h2>

<h3>Einkommen im Vergleich zu Satisfaction</h3>

```python
data = df[['ConvertedCompYearly', 'JobSat']].dropna().sort_values('ConvertedCompYearly')

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(data['ConvertedCompYearly'], data['JobSat'])

ax.set_xlabel('Jahreseinkommen (USD)')
ax.set_ylabel('Jobzufriedenheit')
ax.set_title('Jobzufriedenheit in Abhngigkeit vom Einkommen')

ax.ticklabel_format(style='plain', axis='x') 
ax.set_xlim(0, 1_000_000)

plt.tight_layout()
plt.show()
```

```python
data = df[['ConvertedCompYearly', 'JobSat']].dropna().sort_values('ConvertedCompYearly')
data['JobSat_smooth'] = data['JobSat'].rolling(window=200, center=True).mean()

fig, ax = plt.subplots(figsize=(10,6))

ax.scatter(data['ConvertedCompYearly'], data['JobSat'], s=10, alpha=0.3, label='Einzeldaten')
ax.plot(data['ConvertedCompYearly'], data['JobSat_smooth'], color='red', linewidth=2, label='Trend (geglttet)')

ax.set_xlabel('Jahreseinkommen (USD)')
ax.set_ylabel('Jobzufriedenheit')
ax.set_title('Jobzufriedenheit in Abhngigkeit vom Einkommen')

ax.ticklabel_format(style='plain', axis='x') 
ax.set_xlim(0, 1_000_000)

ax.legend()
plt.tight_layout()
plt.show()
```
