```python
rows = []

csv.field_size_limit(sys.maxsize)

with open("project/survey_results_public.csv", "r", encoding="utf-8-sig") as file:
    reader = csv.DictReader(file)
    reader.fieldnames = [name.strip().replace('"', '') for name in reader.fieldnames]

    for row in reader:
        try:
            # Teil-Daten auswählen: ID, Country, WorkExp, CompTotal, AI-Tool-Nutzung
            rows.append([
                int(row["ResponseId"]),                     # Ganzzahl
                row["MainBranch"],
                row["Age"],
                str(row["EdLevel"]),
                row["Country"],                             # String
                float(row["WorkExp"]) if row["WorkExp"] != "NA" else np.nan,  # Kommazahl
                float(row["CompTotal"]) if row["CompTotal"] != "NA" else np.nan,  # Kommazahl
                #row["AISent"],                              # String
                #float(row["CompTotal"]) > 50000             # Boolean
            ])
        except ValueError:
            continue  # Zeilen mit fehlerhaften Werten überspringen

# In NumPy-Array umwandeln
data = np.array(rows, dtype=object)

print("Shape:", data.shape)
print("Beispielzeile:", data[0])
```

```python
response_id = data[0:20,0]
main_branch = data[0:20,1]
age = data[0:20,2]
ed_level = data[0:20,3]
country = data[0:20,4]
work_exp = data[0:20,5]
comp_total = data[0:20,6]

print(response_id)
print(main_branch)
print(age)
print(ed_level)
print(country)
print(work_exp)
print(comp_total)
```

Filtern Sie ein NumPy-Array basierend auf mindestens 2 Bedingungen auf mindestens 2 Arrays/Spalten

```python
filt_comp_total = comp_total > 60000.0
filt_work_exp = work_exp > 7.0
filt_bachelor = np.array(["bachelor" in str(x).lower() for x in ed_level])

e = response_id[filt_comp_total & filt_work_exp & filt_bachelor]
for i in e:
    print(data[i-1])
```

Berechnen Sie mind. 2 Statistiken auf mind. 2 Arrays

```python
comp_total_mean = round(np.nanmean(comp_total),2)
print("Durchschnittliches Einkommen (unterschiedliche Währungen):",comp_total_mean)

work_exp_mean = round(np.nanmean(work_exp),2)
print("Durchschnittliche Arbeitserfahrung:", work_exp_mean, "Jahre")
```
