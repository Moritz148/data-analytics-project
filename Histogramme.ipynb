```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("stack-overflow-developer-survey-2025/survey_results_public_without_SO.csv")
```

<h2>Überblick über Daten verschaffen, durch die generierung von Histogrammen für ausgewählte Spalten</h2>

<h3>1. MainBranch</h3>

```python
counts = df['MainBranch'].value_counts()

fig, ax = plt.subplots(figsize=(12, 6))

colors = plt.cm.tab10(np.linspace(0, 1, len(counts)))

bars = ax.bar(range(len(counts)), counts.values, color=colors)

ax.set_xlabel('Beruf / Haupttätigkeit')
ax.set_ylabel('Absolute Häufigkeit')
ax.set_title('Verteilung nach Haupttätigkeit')

ax.set_xticks([])

ax.legend(
    bars, counts.index,
    title="Antworten",
    loc='upper center',
    bbox_to_anchor=(0.5, -0.15),
    ncol=2,
    fontsize=9
)

plt.tight_layout()
plt.show()
```
